```python
import psycopg2
import pandas as pd
import json
import app.backend.config as conf
from pathlib import Path
```

```python
# Fonction pour remplacer les valeurs du dictionnaire
def _get_dict_replace():

    replace_natdec = zip(["01", "02", "03", "09"],
                ["Budget primitif", "Décision modificative", "Budget supplémentaire", "Compte administratif"])
    replace_typebud = zip(["P", "A"],
                        ["Principal","Annexe"])
    dict_replace = dict({'nature_dec' : dict(replace_natdec),
                        'type_budget' : dict(replace_typebud)})
    return dict_replace  
```

```python
def get_connexion():
    return psycopg2.connect(database ="actes_budgetaire", user = "postgres",
                        password = "vGXGLqTY4EF7DCiozYoKlqRRUaC7ECd", host = "ab-postgres",
                        port = "5432")

def get_data(connexion, annee, nature_dec):
    query= f"""SELECT * from DocumentBudgetaire
    
    JOIN collectivite 
        ON collectivite.siret_coll = documentbudgetaire.fk_siret_collectivite
    WHERE documentbudgetaire.exercice = '{annee}'
    AND documentbudgetaire.nature_dec = '{nature_dec}' 
    
    """
    data = pd.read_sql_query(query, connexion )
    return data

def get_final_data(interesting_colonnes, df):   
    to_drop = [item for item in list(df.columns) if item not in interesting_colonnes]
    df_final = df.drop(to_drop, axis=1) 
    df_final.replace(_get_dict_replace() , inplace=True)
    return df_final

def get_json_budget(json_columns, df):
    dfd = []
    dict_Lbudget = dict()
    for index, row  in df.iterrows():
        json_df_clean = pd.DataFrame(columns=json_columns)
        dict_Lbudget[index] = json.loads(row["json_budget"])
        for element in dict_Lbudget[index]:
            if element:
                for field in df:
                    element.setdefault(field, None)
        json_df = pd.json_normalize(dict_Lbudget[index])
        dfd.append(json_df.assign(**row.drop("json_budget")))
    data_final = pd.concat(dfd)
    return data_final

def generate_budget_csv(json_columns, df, path: Path, interesting_colonnes):
    dfd = []
    dict_Lbudget = dict()
    for index, row  in df.iterrows():
        dict_Lbudget = json.loads(row["json_budget"])
        for element in dict_Lbudget:
            if element:
                for field in json_columns:
                    element.setdefault(field, None)
        json_df = pd.json_normalize(dict_Lbudget)
        df_temp = json_df.assign(**row.drop("json_budget"))
        if index == 0:
            all_columns = interesting_colonnes + json_columns
            all_columns.remove("json_budget")
            generate_csv(df_temp, path, all_columns)
        else:
            append_csv(df_temp, path, all_columns)

def generate_csv(df, path: Path, annexe_fields):
    return df.to_csv(path, mode='a', index=False, columns=annexe_fields, header=True)

def append_csv(df, path: Path, annexe_fields):
    return df.to_csv(path, mode='a', index=False, columns=annexe_fields, header=False)
```

```python
def export_budget_csv(annee : int, nature_dec : str, nom_csv : str):
    with get_connexion() as connexion:
        df_temporaire = get_data(connexion, annee, nature_dec)
    interesting_colonnes = ['exercice','siret_etablissement', 'libelle', 'nomenclature', 'nature_dec', 'nature_vote','type_budget','json_budget', 'siret_coll', 'libelle_collectivite', 'nature_collectivite']
    df_temporaire = get_final_data(interesting_colonnes, df_temporaire)
    json_column = conf.CHAMPS_LIGNE_BUDGET
    #df_temporaire = get_json_budget(json_column, df_temporaire)
    path = Path.cwd() / "csv" / "2020" / nom_csv
    generate_budget_csv(json_column, df_temporaire, path, interesting_colonnes)
    #generate_csv(df_temporaire, path)

```
